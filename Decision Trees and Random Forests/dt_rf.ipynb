#preparing the data for training
this usually involves few main steps:

*   create train/test/validation split
*   identify input and target columns

*   identify numeric and categorical columns
*   impute(fill) the missing values

*   scale numeric values to 0,1
*   Encode categorical columns to one-hot vectors







##Training, Validation and Test Sets

```python
plt.title("number of raws per year")
sns.countplot(x=pd.to_datetime(raw_df.Date).dt.year)
```

While working with chronological data, it's often a good idea to separate the training, validation and test sets with time, so that the model is trained on data from the past and evaluated on data from the future.

We'll use the data till 2014 for the training set, data from 2015 for the validation set, and the data from 2016 & 2017 for the test set.


```python
year=pd.to_datetime(raw_df.Date).dt.year
train_data=raw_df[year<2014]
validation_data=raw_df[year==2015]
test_data=raw_df[year>2015]
```

```python
train_data
```

```python
train_data.shape,validation_data.shape,test_data.shape
```

##identifying input and target columns

```python
#now lets set input and target columns
input_cols=list(raw_df.columns)[1:-1]
target_cols="RainTomorrow"
```

```python
train_input=train_data[input_cols]
train_target=train_data[target_cols]
```

```python
validation_input=validation_data[input_cols]
validation_target=validation_data[target_cols]
```

```python
test_input=test_data[input_cols]
test_target=test_data[target_cols]
```

```python
train_input.columns
```

```python
train_target
```

##identifying numeric and categorical columns

```python
numeric_cols=train_input.select_dtypes(include=np.number).columns.tolist()
categorical_cols=train_input.select_dtypes(exclude=np.number).columns.tolist()
```

```python
numeric_cols,categorical_cols
```

```python
# EXERCISE: Study how various columns are correlated with the target and select just a subset of the columns, instead of all of the. Observe how it affects the results
```

##imputation

```python
train_input.isnull().sum()
```

```python
#now lets replace the null values of numeric columns with mean value
from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy="mean").fit(raw_df[numeric_cols])
```

```python
train_input[numeric_cols]=impute.transform(train_input[numeric_cols])
validation_input[numeric_cols]=impute.transform(validation_input[numeric_cols])
test_input[numeric_cols]=impute.transform(test_input[numeric_cols])
```

##scaling upto 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler().fit(raw_df[numeric_cols])
```

```python
train_input[numeric_cols]=scale.transform(train_input[numeric_cols])
validation_input[numeric_cols]=scale.transform(validation_input[numeric_cols])
test_input[numeric_cols]=scale.transform(test_input[numeric_cols])
```

```python
train_input[numeric_cols]
```

##encoding the categorical data to one hot vector

```python
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder(sparse_output=False, handle_unknown="ignore").fit(raw_df[categorical_cols])
```

```python
encoded_cols=list(encode.get_feature_names_out(categorical_cols))
```

```python
train_input[encoded_cols]=encode.transform(train_input[categorical_cols])
validation_input[encoded_cols]=encode.transform(validation_input[categorical_cols])
test_input[encoded_cols]=encode.transform(test_input[categorical_cols])
```

```python
train_input
```

```python
#final step
x_train=train_input[numeric_cols+encoded_cols]

x_validation=validation_input[numeric_cols+encoded_cols]

x_test=test_input[numeric_cols+encoded_cols]
```

```python
x_test
```

#training and visualizing the decision tree

##training

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=42)
model.fit(x_train,train_target)
```

##evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix
train_preds=model.predict(x_train)
train_preds
```

```python
pd.value_counts(train_preds)
```

```python
train_prob=model.predict_proba(x_train)
train_prob
```

now lets check the accuracy

```python
accuracy_score(train_preds,train_target)
```

The training set accuracy is close to 100%! But we can't rely solely on the training set accuracy, we must evaluate the model on the validation set too.

We can make predictions and compute accuracy in one step using model.score

```python
model.score(x_validation,validation_target)
```

Although the training accuracy is 100%, the accuracy on the validation set is just about 79%, which is only marginally better then always predicting "No".

```python
validation_target.value_counts()/len(validation_target)
```

#visualization


```python
from sklearn.tree import plot_tree, export_text
plt.figure(figsize=(80,20))
plot_tree(model,feature_names=x_train.columns,max_depth=2,filled=True)
```

```python
#now lets check the max depth of the tree
model.tree_.max_depth
```

Based on the gini index computations, a decision tree assigns an "importance" value to each feature. These values can be used to interpret the results given by a decision tree.

```python
model.feature_importances_
```

```python
#now lets view it in dataframe
important_df=pd.DataFrame(
    {
        "features":x_train.columns,
        "importance":model.feature_importances_
    }
).sort_values('importance',ascending=False)
```
