##Training, Validation and Test Sets

```python
plt.title("number of raws per year")
sns.countplot(x=pd.to_datetime(raw_df.Date).dt.year)
```

While working with chronological data, it's often a good idea to separate the training, validation and test sets with time, so that the model is trained on data from the past and evaluated on data from the future.

We'll use the data till 2014 for the training set, data from 2015 for the validation set, and the data from 2016 & 2017 for the test set.


```python
year=pd.to_datetime(raw_df.Date).dt.year
train_data=raw_df[year<2014]
validation_data=raw_df[year==2015]
test_data=raw_df[year>2015]
```

```python
train_data
```

```python
train_data.shape,validation_data.shape,test_data.shape
```

##identifying input and target columns

```python
#now lets set input and target columns
input_cols=list(raw_df.columns)[1:-1]
target_cols="RainTomorrow"
```

```python
train_input=train_data[input_cols]
train_target=train_data[target_cols]
```

```python
validation_input=validation_data[input_cols]
validation_target=validation_data[target_cols]
```

```python
test_input=test_data[input_cols]
test_target=test_data[target_cols]
```

```python
train_input.columns
```

```python
train_target
```

##identifying numeric and categorical columns

```python
numeric_cols=train_input.select_dtypes(include=np.number).columns.tolist()
categorical_cols=train_input.select_dtypes(exclude=np.number).columns.tolist()
```

```python
numeric_cols,categorical_cols
```

```python
# EXERCISE: Study how various columns are correlated with the target and select just a subset of the columns, instead of all of the. Observe how it affects the results
```

##imputation

```python
train_input.isnull().sum()
```

```python
#now lets replace the null values of numeric columns with mean value
from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy="mean").fit(raw_df[numeric_cols])
```

```python
train_input[numeric_cols]=impute.transform(train_input[numeric_cols])
validation_input[numeric_cols]=impute.transform(validation_input[numeric_cols])
test_input[numeric_cols]=impute.transform(test_input[numeric_cols])
```

##scaling upto 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler().fit(raw_df[numeric_cols])
```

```python
train_input[numeric_cols]=scale.transform(train_input[numeric_cols])
validation_input[numeric_cols]=scale.transform(validation_input[numeric_cols])
test_input[numeric_cols]=scale.transform(test_input[numeric_cols])
```

```python
train_input[numeric_cols]
```

##encoding the categorical data to one hot vector

```python
from sklearn.preprocessing import OneHotEncoder
encode=OneHotEncoder(sparse_output=False, handle_unknown="ignore").fit(raw_df[categorical_cols])
```

```python
encoded_cols=list(encode.get_feature_names_out(categorical_cols))
```

```python
train_input[encoded_cols]=encode.transform(train_input[categorical_cols])
validation_input[encoded_cols]=encode.transform(validation_input[categorical_cols])
test_input[encoded_cols]=encode.transform(test_input[categorical_cols])
```

```python
train_input
```

```python
#final step
x_train=train_input[numeric_cols+encoded_cols]

x_validation=validation_input[numeric_cols+encoded_cols]

x_test=test_input[numeric_cols+encoded_cols]
```

```python
x_test
```

#hyper parameter tuning

As we saw in the previous section, our decision tree classifier memorized all training examples, leading to a 100% training accuracy, while the validation accuracy was only marginally better than a dumb baseline model. This phenomenon is called overfitting, and in this section, we'll look at some strategies for reducing overfitting.

The DecisionTreeClassifier accepts several arguments, some of which can be modified to reduce overfitting.

These arguments are called hyperparameters because they must be configured manually (as opposed to the parameters within the model which are learned from the data. We'll explore a couple of hyperparameters:

    max_depth
    max_leaf_nodes

max_depth

By reducing the maximum depth of the decision tree, we can prevent the tree from memorizing all training examples, which may lead to better generalization

```python
model=DecisionTreeClassifier(max_depth=3,random_state=42)
```

```python
model.fit(x_train,train_target)
```

```python
model.score(x_train,train_target)
```

```python
model.score(x_validation,validation_target)
```

Great, while the training accuracy of the model has gone down, the validation accuracy of the model has increased significantly.

```python
model.classes_
```

```python
plt.figure(figsize=(80,20))
plot_tree(model, feature_names=x_train.columns,filled=True, rounded=True, class_names=model.classes_)
```

Let's experiment with different depths using a helper function.

```python
def max_depth_eror(md):
  model=DecisionTreeClassifier(max_depth=md, random_state=42)
  model.fit(x_train,train_target)
  train_acc=1-model.score(x_train,train_target)
  validation_acc=1-model.score(x_validation,validation_target)
  return {"max depth":md,
          "train acc":train_acc,
          "validation_acc":validation_acc}
```

```python
error_df=pd.DataFrame([max_depth_eror(md) for md in range(1,21)])
```

```python
error_df
```

in validation col the acc decreases and after certain points , it starts increasing again


```python
plt.figure()
plt.plot(error_df['max depth'],error_df['train acc'])

plt.plot(error_df['max depth'],error_df['validation_acc'])
plt.xticks(range(0,21, 2))
plt.xlabel("max_depth")
plt.ylabel("prediction error(1-accuracy)")
plt.legend(["training","validation"])
```

```python
model = DecisionTreeClassifier(max_depth=7, random_state=42).fit(x_train, train_target)
model.score(x_validation, validation_target)
```

max_leaf_nodes

Another way to control the size of complexity of a decision tree is to limit the number of leaf nodes. This allows branches of the tree to have varying depths.

```python
model=DecisionTreeClassifier(max_leaf_nodes=128,random_state=42)
model.fit(x_train,train_target)
```

```python
model.score(x_train,train_target)
```

```python
model.score(x_validation,validation_target)
```

```python
model.tree_.max_depth
```

Notice that the model was able to achieve a greater depth of 12 for certain paths while keeping other paths shorter.



#random forest

While tuning the hyperparameters of a single decision tree may lead to some improvements, a much more effective strategy is to combine the results of several decision trees trained with slightly different parameters. This is called a random forest model.

The key idea here is that each decision tree in the forest will make different kinds of errors, and upon averaging, many of their errors will cancel out. This idea is also commonly known as the "wisdom of the crowd":

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_jobs=-1,random_state=42)
model.fit(x_train,train_target)
```

```python
model.score(x_train,train_target)
```

```python
model.score(x_validation,validation_target)
```

Once again, the training accuracy is almost 100%, but this time the validation accuracy is much better. In fact, it is better than the best single decision tree we had trained so far. Do you see the power of random forests?

This general technique of combining the results of many models is called "ensembling", it works because most errors of individual models cancel out on averaging.

```python
train_prob=model.predict_proba(x_train)
```

Just like decision tree, random forests also assign an "importance" to each feature, by combining the importance values from individual trees.
